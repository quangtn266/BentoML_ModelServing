```python
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import numpy as np
```

```python
predictions = [350, 430, 550 ,300]
actual_jan = [350, 430, 550, 300]
actual_feb = [360, 445, 570, 305]
actual_mar = [370, 460, 590, 310]
actual_apr = [380, 475, 610, 315]
actual_may = [390, 490, 630, 320]
actual_jun = [400, 505, 650, 335]
actual_july = [430, 530, 670, 340]
```

```python
mse_jan = mean_squared_error(actual_jan, predictions)
mse_feb = mean_squared_error(actual_feb, predictions)
```

```python
mse_mar = mean_squared_error(actual_mar, predictions)
mse_apr = mean_squared_error(actual_apr, predictions)
mse_may = mean_squared_error(actual_may, predictions)
mse_jun = mean_squared_error(actual_jun, predictions)
mse_july = mean_squared_error(actual_july, predictions)
```

```python
errors = np.array([mse_jan, mse_feb, mse_mar, mse_apr, mse_may, mse_jun, mse_july])
```

```python
plt.plot(errors)
plt.ylabel("MSE")
plt.xlabel("Index of Month")
plt.show()
```

```python
import math

MSE = np.array([0, 4000, 7000, 11000, 15000, 23000])
MAE = np.array([0, 200, 300, 500, 700, 1000])
MAPE = np.array([0, 3, 8, 12, 20, 30])
RMSE = np.array([0, math.sqrt(4000), math.sqrt(7000),
                        math.sqrt(11000), math.sqrt(15000),math.sqrt(23000)])
```

```python
fig, ax = plt.subplots(2, 2)
fig.suptitle("Model monitoring dashboard")
ax[0, 0].plot(MSE)
ax[0, 0].set(xlabel="Index of Month", ylabel="MSE")
ax[0, 1].plot(MAE)
ax[0, 1].set(xlabel="Index of Month", ylabel="MAE")
ax[1, 0].plot(MAPE)
ax[1, 0].set(xlabel="Index of Month", ylabel="MAPE")
ax[1, 1].plot(RMSE)
ax[1, 1].set(xlabel="Index of Month", ylabel="RMSE")
plt.show()
```

```python
MSE_Thresholds = np.array([5000, 5000, 5000, 5000, 5000, 5000])
fig, ax = plt.subplots(2, 2)
fig.suptitle("Model monitoring dashboard")
ax[0, 0].plot(MSE)
ax[0, 0].plot(MSE_Thresholds, 'r--', label="MSE Threshold")
ax[0, 0].legend()
ax[0, 0].set(xlabel="Index of Month", ylabel="MSE")
```
